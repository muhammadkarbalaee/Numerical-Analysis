```python
from sympy import fourier_series, pi, plot
from sympy.abc import x
f = x
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 2)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 10)
p = plot(f, s1, s2, s3, (x, -2, 2), show=False, legend=True)
```

```python
p[0].line_color = (0, 0, 0)
p[0].label = 'x'
p[1].line_color = (1, 0, 0.7)
p[1].label = 'n=2'
p[2].line_color = (0.5, 0.5, 1)
p[2].label = 'n=5'
p[3].line_color = (0, 1, 1)
p[3].label = 'n=10'
```

```python
p.show()
```
