The implementation of the formula above is defined down below.

```python
def trapezoid_integral(bar_results:list,h:int)->int:
    result = 0.0000
    for i in range(len(bar_results)):
        if(i==0 or i==(len(bar_results)-1)):
            result = np.add(bar_results[i],result) 
        result = np.add(np.multiply(2,bar_results[i]),result) 
    return np.multiply(np.divide(h,2),result)
```

```python
def bar_calculation(a:int,b:int,n:int,h:int,percision:int)->list:
    num_to_cal = np.linspace(0,1,n+1)
    bar_results = []
    for i in range(n+1):
        f_x = raw_function.subs(x,num_to_cal[i]).evalf(percision)
        bar_results.append(f_x)
    return bar_results    
```

```python
raw_function = sp.exp(-pow(x,2))
a = 0
b = 1 
n = 10000
h = (b-a)/n
percision = 50
```

```python
raw_function
```

```python
bar_results = bar_calculation(a=a,b=b,n=n,h=h,percision=percision)
```

```python
result = trapezoid_integral(bar_results=bar_results,h=h)
print("The result for {} n is {}".format(n,result))
correct_answer = raw_function.integrate(x).subs(x,1).evalf(percision)- raw_function.integrate(x).subs(x,0).evalf(percision)
print('The correct answer is {}'.format(correct_answer))
print('The diffrence between the correct and predicted anser is {}'.format(np.subtract(result,correct_answer)))
```
