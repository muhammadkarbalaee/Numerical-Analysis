# 2nd Question

```python
x = sym.Symbol('x')
f = cos(2*x)
fprim =  Derivative(f)

a0 = sym.Symbol('a0')
a1 = sym.Symbol('a1')
a2 = sym.Symbol('a2')
a3 = sym.Symbol('a3')
a4 = sym.Symbol('a4')
a5 = sym.Symbol('a5')
a6 = sym.Symbol('a6')
a7 = sym.Symbol('a7')


p = a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4 + a5*x**5 + a6*x**6 + a7*x**7
pprim = p.diff(x)

points = [0.2,0.4,0.6,0.8]
```

```python
B = []
for i in points:
    B.append(float(f.subs(x,i).evalf()))
```

```python
for i in points:
    B.append(float(fprim.subs(x,i).evalf()))
```

```python
A = []
for i in points:
    temp = []
    for j in [a0,a1,a2,a3,a4,a5,a6,a7]:
        temp.append(float(p.subs(x,i).coeff(j)))
    A.append(temp)
```

```python
# coeffs = []
for i in points:
    temp = []
    for j in [a0,a1,a2,a3,a4,a5,a6,a7]:
        temp.append(float(pprim.subs(x,i).coeff(j)))
    A.append(temp)
```

```python
A
```

```python
y = np.linalg.solve(np.array(A),np.array(B))
```

```python
y
```

```python
from sympy import symbols
from sympy.plotting import plot
x = symbols('x')
p1 = plot(y[0]+y[1]*x+y[2]*x*x+y[3]*x**3+y[4]*x**4+y[5]*x**5+y[6]*x**6+y[7]*x**7,(x, 0, 1), show=True)
p2 = plot(cos(2*x),(x,0, 1), show=True)
p3 = plot(cos(2*x)-(y[0]+y[1]*x+y[2]*x*x+y[3]*x**3+y[4]*x**4+y[5]*x**5+y[6]*x**6+y[7]*x**7),(x,0, 1), show=True)
p1.append(p2[0])
p1.show()
```

# Third Question 

```python
x = sym.Symbol('x')

a0 = sym.Symbol('a0')
a1 = sym.Symbol('a1')
a2 = sym.Symbol('a2')
a3 = sym.Symbol('a3')
a4 = sym.Symbol('a4')
a5 = sym.Symbol('a5')
a6 = sym.Symbol('a6')


p = a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4 + a5*x**5 + a6*x**6 
pprim = p.diff(x)
pprimprim = pprim.diff(x)

points = [2,4,5]
B = [12,18,13,0,1,0,-0.7,-0.7,-0.7]
```

```python
A = []
for i in points:
    temp = []
    for j in [a0,a1,a2,a3,a4,a5,a6]:
        temp.append(float(p.subs(x,i).coeff(j)))
    A.append(temp)
```

```python
# coeffs = []
for i in points:
    temp = []
    for j in [a0,a1,a2,a3,a4,a5,a6]:
        temp.append(float(pprim.subs(x,i).coeff(j)))
    A.append(temp)
```

```python
# coeffs = []
for i in points:
    temp = []
    for j in [a0,a1,a2,a3,a4,a5,a6]:
        temp.append(float(pprimprim.subs(x,i).coeff(j)))
    A.append(temp)
```

 As we know for solving the Ax=B problem with commom method that A matrix should be the square but here it is not and we should use other methods to solve 
 system of equations using matrix methods.

```python
A
```

```python
# This method will not work
#y = np.linalg.solve(np.array(A),np.array(B))
```

```python
#Instead we use this method which can solve non-square matrix equations
y = np.linalg.lstsq(A, B, rcond=None)[0]
```
