## Problem 1
Plot each one of these functions with their error in an appropriate interval:


### Part a)
$sin(x) , x$

```python
from sympy import symbols
from sympy.plotting import plot
from sympy import *
x = symbols('x')
```

```python
def plot_function(function ):
    p1 = plot(function , (x,-2,2),show=False)
    p1.show()
```

```python
function = sin(x)
approximation = x 
error_function = function - approximation

plot_function(function=function)
plot_function(function=approximation)
plot_function(function=error_function)
```

### Part b)
$sin(x) , x - (x^3)/3!$

```python
function = sin(x)
approximation = x - (pow(x,3)/factorial(3))
error_function = function - approximation

plot_function(function=function)
plot_function(function=approximation)
plot_function(function=error_function)
```

## Problem 2 

The Taylor series expansion for the cosine function is:

$$\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!}x^{2n} = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!} + \cdots$$

To construct a polynomial that approximates the cosine function with an error lower than $10^{-7}$, we can use the first few terms of the Taylor series expansion. Specifically, if we use the first $n$ terms of the series, the error in approximating the cosine function will be no larger than the $(n+1)$th term of the series.
To determine the number of terms we need to use in the series, we can solve for $n$ in the inequality:

$$\frac{x^{n+1}}{(n+1)!} < 10^{-7}$$

Taking the logarithm of both sides and using Stirling's approximation for factorials, we get:

$$\ln(x) - \ln(n+1) - \frac{1}{2}\ln(2\pi) - (n+1)\ln(10) < -\ln(10^7)$$

Simplifying and solving for $n$, we obtain:

$$n > \frac{\ln(x) - \ln(\sqrt{2\pi} \cdot 10^3) + \ln(\ln(10^7))}{\ln(x) - \ln(\ln(10))} - 1$$

For $x \leq 1$, we can use $n = 4$ to obtain an approximation with an error less than $10^{-7}$. Thus, our polynomial approximation for $\cos(x)$ is:

$$p(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!} + \frac{x^8}{8!}$$

For $x > 1$, we would need to use a larger number of terms in the series to achieve an approximation with an error less than $10^{-7}$. However, for the purposes of this question, we will only consider values of $x$ less than or equal to 1.

Now let's see how it works in practice &#x1F600;


```python
function = cos(x)
approximation = 1 - (pow(x,2)/2) + pow(x,4)/factorial(4) - pow(x,6)/factorial(6) + pow(x,8)/factorial(8)
error_function = function - approximation

plot_function(function=function)
plot_function(function=approximation)
plot_function(function=error_function)
```
